```python
key = ["Vegetables","veggies","Avacado",'Meat','Potato','Onion','Chicken','Apple','Melon','Bread','Cheese','Tomato','Fruit','Bagel','Orange','Lemon','Burger','Pepper','Steak','Lettuce','Carrot','Corn','Cucumber','Turkey','Mango','Lime','Ham','Cabbage','Sausage','Coconut','Pork','Brownies','Garlic','Cake','Bacon','Wine Bottle','Pizza','Sandwich','Oyesters','Pineapple','Muffin','Hot Dog','Cauliflower','Brussel Sprouts','Beef','Roast','Strawberries','Mushroom','Grape','Banana','Yams','Salmon','Kiwi','Kale','Ice cream','Herbs','Beets','Pie','Lobster','Greens','Grapefruit','Goose','Clams','Champagne','Broccoli','Yucca','Shrimp','Papusa','Duck']
key = [x.upper() for x in key]
a = dict.fromkeys(key,0)
```

```python
for n in g1['Narrative_1']:
    for word in key:
        if(n.find(word)>=0):
            a[word]+=1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a['other'] = 0
for v in a:
    if(a.get(v)<15):
        a['other'] += a.get(v)
        a[v]=0
af = {x:y for x,y in a.items() if y!=0}
```

```python
plt.figure(figsize=(15,15))
x=[float(v) for v in af.values()]
y=[k for k in af]
plt.pie(x,labels=y )
plt.show()
```

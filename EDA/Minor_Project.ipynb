# Minor Project


```python
import pandas as pd
scores=pd.read_csv('performance.csv')
scores1=scores.loc[:,'math score':'writing score']
scores['avg_score']=scores1.mean(axis=1)
```

importing the csv file and converting it to a dataframe using pandas as there are three scores mentioned in the dataset 
an average of the three scores is added in dataframe for an observation 

```python
scores.head(5)
```

```python
race=scores.groupby('race/ethnicity')
gender=scores.groupby('gender')
education=scores.groupby('parental level of education')
lunch=scores.groupby('lunch')
prep=scores.groupby('test preparation course')
```

The parameters which are mentioned in dataset are namely on the basis of 
* race
* gender
* education
* food before the exam
* prepration of the candidate

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
parameter='avg_score'
```

## Analysis of the Average Score 

```python
f, axes = plt.subplots(2, 3, figsize=(14, 8), sharex=True,squeeze=False)
sns.distplot(race.get_group('group A')[parameter], ax=axes[0,0])
axes[0,0].set_title('Group A')
sns.distplot(race.get_group('group B')[parameter], ax=axes[0,1])
axes[0,1].set_title('Group B')
sns.distplot(race.get_group('group C')[parameter], ax=axes[0,2])
axes[0,2].set_title('Group C')
sns.distplot(race.get_group('group D')[parameter], ax=axes[1,0])
axes[1,0].set_title('Group D')
sns.distplot(race.get_group('group E')[parameter], ax=axes[1,1])
axes[1,1].set_title('Group E')
f.delaxes(axes[1,2])
f.suptitle("Comparision between all races")
```

```python
for name,group in race:
    sns.distplot(race.get_group(name)[parameter],kde_kws={"label":name},hist=False)
plt.figure(figsize=(20,10))
for name,group in race:
    print("The minimum value in the Race "+name+" is "+str(round(race.get_group(name).avg_score.min()))+" and the maximum is "+str(round(race.get_group(name).avg_score.max())))
```

It is evident that the candidates of Group A have not performed good in the test as the density of the scores as compared to the other candidates is less 
Group E has performed the best among all the group as their density of the marks is the best eventough the maximum marks were scored by student in Group D


```python
f,axes=plt.subplots(1,3,figsize=(9, 4), sharex=True,squeeze=False)
sns.distplot(gender.get_group('male')[parameter],ax=axes[0,0])
axes[0,0].set_title("Males")
sns.distplot(gender.get_group('female')[parameter],ax=axes[0,1])
axes[0,1].set_title('Females')
f.suptitle("Gender comparision")
for name,group in gender:
    sns.distplot(gender.get_group(name)[parameter],ax=axes[0,2],kde_kws={"label":name})
for name,group in gender:
    print("The minimum value for the gender "+name+" is "+str(round(gender.get_group(name).avg_score.min()))+" and the maximum is "+str(round(gender.get_group(name).avg_score.max())))    
```

It is clear from the above graphs the Females have outperformed males in the test both in the maximum marks and the overall marks density

```python
f,axes=plt.subplots(2,3,figsize=(14, 8), sharex=True,squeeze=False)
sns.distplot(education.get_group("associate's degree")[parameter],ax=axes[0,0])
axes[0,0].set_title("associate's degree")
sns.distplot(education.get_group("bachelor's degree")[parameter],ax=axes[0,1])
axes[0,1].set_title("bachelor's degree")
sns.distplot(education.get_group("high school")[parameter],ax=axes[1,0])
axes[1,0].set_title("high school")
sns.distplot(education.get_group("master's degree")[parameter],ax=axes[1,1])
axes[1,1].set_title("master's degree")
sns.distplot(education.get_group("some college")[parameter],ax=axes[0,2])
axes[0,2].set_title("some college")
sns.distplot(education.get_group("some high school")[parameter],ax=axes[1,2])
axes[1,2].set_title("some high school")
f.suptitle("Comparision of academic qualifications")
```

```python
for name,group in education:
    sns.distplot(education.get_group(name)[parameter],kde_kws={"label":name},hist=False)
plt.figure(figsize=(20,10))
for name,group in education:
    print("The minimum value of people with "+name+" is "+str(round(education.get_group(name).avg_score.min()))+" and the maximum is "+str(round(education.get_group(name).avg_score.max())))
```

There is nothing clear from this data parameter except that people with masters degree have better marks density

```python
f,axes=plt.subplots(1,3,figsize=(10, 6), sharex=True,squeeze=False)
sns.distplot(prep.get_group('completed')[parameter],ax=axes[0,0])
axes[0,0].set_title("completed")
sns.distplot(prep.get_group('none')[parameter],ax=axes[0,1])
axes[0,1].set_title('none')
f.suptitle("Preperation done by candidate")
for name,group in prep:
    sns.distplot(prep.get_group(name)[parameter],ax=axes[0,2],kde_kws={"label":name})
for name,group in prep:
    print("The minimum value for people with "+name+" preperation "+" is "+str(round(prep.get_group(name).avg_score.min()))+" and the maximum is "+str(round(prep.get_group(name).avg_score.max())))
    
```

```python
f,axes=plt.subplots(1,3,figsize=(10, 6), sharex=True,squeeze=False)
sns.distplot(lunch.get_group('free/reduced')[parameter],ax=axes[0,0])
axes[0,0].set_title("free/reduced")
sns.distplot(lunch.get_group('standard')[parameter],ax=axes[0,1])
axes[0,1].set_title('standard')
f.suptitle("Lunch")
for name,group in lunch:
    sns.distplot(lunch.get_group(name)[parameter],ax=axes[0,2],kde_kws={"label":name})
for name,group in lunch:
    print("The minimum value of people with "+name+" Lunch "+" is "+str(round(lunch.get_group(name).avg_score.min()))+" and the maximum is "+str(round(lunch.get_group(name).avg_score.max())))
```
